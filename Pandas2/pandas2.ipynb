```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    df = pd.read_csv('crime_data.csv')

    # create plot of crime frequency ####################################################
    df.plot(kind='hist',y=["Total"], bins=8,title="Frequency of total crime values")
    plt.xlabel("Total Number of crimes")
    plt.show()
    print("The above plot demonstrates the range of different total crime values per year, and denotes their frequency")
    
    # create comparison of violent vs nonviolent crimes #################################
    df["Nonviolent"] = df.Total-df.Violent
    df.plot(x= 'Year',y=["Violent",'Nonviolent'],title="Frequency of violent and non-violent crimes by year")
    plt.xlabel("Number of Crimes")
    plt.tight_layout()
    plt.show()
    print("The above trends show that violent and non-violent crime rates are not static across time")
    
    # create plot of crimes per population ##############################################
    df['Percentage'] = df.Total/df.Population
    df.plot(x='Year', y='Percentage', title="Total crime as a percentage of total population")
    plt.ylabel("Percentage")
    plt.show()
    
    print("This curve is similar to the trends in the previous graph, showing that crime variance is not uniquely driven by population size")
    
    
    
    
prob1()
```

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Thefts against the number of Robberies
    """
    df = pd.read_csv('crime_data.csv')
   
    df.plot(kind='hexbin',x="Vehicle Theft", y="Robbery", gridsize=15, sharex=False, title="A comparisone of rates of different crime types")
    plt.show()

    df.plot(kind='box', y = ['Burglary','Violent', 'Vehicle Theft'], title="A comparison of the distribution across years for different types of crimes\n")
    plt.tight_layout()
    plt.show()

prob3()
```
